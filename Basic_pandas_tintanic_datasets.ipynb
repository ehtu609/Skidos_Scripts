# Selecting values which arre not null

```python
age_no_na=df[df["Age"].notna()]
age_no_na
```

# The same below class 2_3 can be selected with or command '|' 

```python
class_2_3=df[(df["Pclass"]==2) | (df["Pclass"]==3)]
class_2_3.head()
```

# If intererested in two values of a single column like here for passengers in class 2 and 3 

```python
class_2_3=df[df["Pclass"].isin([2,3])]
```
